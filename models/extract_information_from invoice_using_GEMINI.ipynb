```python
from pathlib import Path

def image_format(image_path):
    img = Path(image_path)

    if not img.exists():
        raise FileNotFoundError(f"Could not find image: {img}")

    image_parts = [
        {
            "mime_type": "image/png", ## Mime type are PNG - image/png. JPEG - image/jpeg. WEBP - image/webp
            "data": img.read_bytes()
        }
    ]
    return image_parts


image_format('/home/manish/Desktop/projects/etfp/bill_image_receipt.png')
```

```python
def gemini_output(image_path,
                  system_prompt,
                  user_prompt):
    
    image_info = image_format(image_path)
    print(image_info[0])
    input_prompt = [system_prompt, image_info[0], user_prompt]
    response = model.generate_content(input_prompt)

    return response.text


```

```python
import google.generativeai as genai 

google_api = 'AIzaSyAMlYqwvuQgekl8nlqc56XTqJVBufszrBU'
genai.configure(api_key=google_api)

# Model Configuration
MODEL_CONFIG = {
  "temperature": 0.2,
  "top_p": 1,
  "top_k": 32,
  "max_output_tokens": 4096,
}

## Safety Settings of Model
safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  }
]

model = genai.GenerativeModel(
                            # model_name='gemini-2.5-flash',  # 6 min takes to 4 image 
                            model_name='gemini-2.5-pro',
                              generation_config=MODEL_CONFIG,
                              safety_settings=safety_settings)
model



def gemini_output(image_path,
                  system_prompt,
                  user_prompt):
    
    
    input_prompt = [system_prompt, image_path, user_prompt]
    response = model.generate_content(input_prompt)

    return response.text


```

```python
def process_multiple_images(image_paths,
                            system_prompt,
                            user_prompt):

    all_results = []

    for image_path in image_paths:
        result = gemini_output(image_path, system_prompt, user_prompt)
        all_results.append({
            result
        })

    return all_results
    

```
